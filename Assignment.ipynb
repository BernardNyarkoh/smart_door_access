```python
class Staff:
    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code

    # Getter using @property
    @property
    def access_code(self):
        auth = input("Enter admin password to view access code: ")
        if auth == "admin123":
            return self.__access_code
        else:
            return "Access Denied!"

    # Setter using @property
    @access_code.setter
    def access_code(self, new_code):
        if len(new_code) >= 6:
            self.__access_code = new_code
            print("Access code updated successfully.")
        else:
            print("Error: Access code must be at least 6 characters long.")

    # Display staff information
    def display_info(self):
        print("Staff Name:", self.s_name)


# ---------------- Driver Code ----------------

staff1 = Staff("Kelvin Acheampong", "UMAT2025")

staff1.display_info()

# View access code
print("Access Code:", staff1.access_code)

# Update access code
staff1.access_code = "NewCode123"

# View updated access code
print("Updated Access Code:", staff1.access_code)
```
